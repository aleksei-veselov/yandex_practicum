#### Обработка колонки rating

**Интерпритация рейтинга ESRB (ESRB была основана в 1994 году):**
- **EC** — «Для детей младшего возраста»: Игра подходит для детей от 3 лет.

- **E** — «Для всех»: Содержание вполне подходит для всех возрастов. Первоначально "K-A".

- **E10+** — «Для всех от 10 лет и старше». Прнят с 02 марта 2005 г.

- **T** — «Подросткам»: Игра подходит для лиц от 13 лет.

- **M** — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет.

- **AO**  — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет.

- **RP**  — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу

- **RP 17+** — «Рейтинг ожидается - примерно 17+»: Продукт отправлен ESRB на оценку, но примерное содержание близко к рейтингу «M». Данная маркировка была введена с 2021 г. и распространяется на демо или презентационный вариант.

Т.к. кол-во пропусков rating составляет 6766, попробуем выявить закономерность пропусков.

```python
nan_rating = df[df['rating'].isna()]
nan_rating
```

Т.к. процессом присвоения рейтинга занимаются многие люди разных возрастов и слоёв общества, мы не сможем корректно заменить пропущенные значения, но при анализе интерпритации рейтинга ESRB выявлено, что встречается тип "K-A", которые в дальнейшем был изменен на тип "Е". Заменим тип "K-A" на "Е". Остальные пропуски заменим на синтетический тип "NA" и в дальнейшем это учитывать при дальнейшем анализе.

Проводя исследование выпуска видеоигр по годам, сделали вывод о том, что данными до 1994 года можно пренебречь.

```python
df = df.loc[df['year_of_release'] >= 1994] # Ограничим значения года выпуска игры до 1994 г.
```

Рассмотрим общие продажи продажи по платформам.

```python
 #группировка по суммарным продажам платформ, году выпуска
platform_total_sales = df.pivot_table(
    index='platform',
    values='total_sales',
    aggfunc='sum'
).sort_values(
    by='total_sales',
    ascending = False
)
platform_total_sales = platform_total_sales.reset_index()
```

Для дальнейшего анализа будем рассматривать первые 10 платформ с наибольшими суммарными продажами.

```python
top_platform_total_sales = platform_total_sales['platform'].head(10)
top_platform_total_sales
```

Для дальнейшего анализа, построим распределение по годам 10 платформ с наибольшими суммарными продажами.

```python

df[df['platform'].isin(top_platform_total_sales)].pivot_table(
    index='year_of_release',
    columns = 'platform',
    values='total_sales',
    aggfunc='sum'
).plot(
    figsize = (10,6),
    kind = 'bar', #тип графика
    stacked = True, #отображение всех значений columns по каждому году
    width = 0.85, # отношение размера расстояния между столбцами к суммарной оси х
    rot = 45); #угол наклона подписи оси х
```

Рассмотрим распределение по годам каждой платформы.

```python
df[df['platform'].isin(top_platform_total_sales)].pivot_table(
    index='year_of_release',
    columns = 'platform',
    values='total_sales',
    aggfunc='sum'
).plot(
    figsize = (10,15),
    kind = 'bar', #тип графика
    stacked = True, #суммирование всех зачений для каждого года
    width = 0.85, # отношение размера расстояния между столбцами к суммарной оси х
    rot = 45, # угол наклона подписей оси х
    subplots = True, # отображение в пределах одной оси х всех columns
    title=['']*10, # убрать название subplots
    grid= True);
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
from scipy import stats as st
```

Для проверки гипотезы о равенстве среднего двух генеральных совокупностей в Python есть специальный тест scipy.stats.ttest_ind (array1, array2, equal_var), где

- rray1, array2 — массивы, содержащие выборки;
- equal_var (от англ. equal variance — «равная дисперсия») — необязательный параметр, задающий, считать ли равными дисперсии выборок. Передаётся как equal_var = True или equal_var = False (True — считать, False — не считать).

Для оценки гипотез применяется t-тест. При проверке гипотез может получиться два ответа "Отвергаем нулевую гипотезу" и "Не получилось отвергнуть нулевую гипотезу".

При ответе "Отвергаем нулевую гипотезу" значит наше предположение неверно. Если получить ответ "Не получилось отвергнуть нулевую гипотезу", это значит что предположение верно.

Т.к. наши генеральные совокупности не велики и составляют менее 200 значений, то критический уровень статистической значимости (alpha) примем =  0.05.

**Формулировка гипотез:**

H_0: Средние пользовательских рейтингов (user_score) платформы Xbox One  = cредним пользовательских рейтингов (user_score) платформы PC

H_1: Средние пользовательских рейтингов (user_score) платформы Xbox One  ≠ cредним пользовательских рейтингов (user_score) платформы PC

**Избавим значения 'user_score' от заглушек**

```python
u = df.loc[df['user_score'] != -1]
```

Определим необходимые значения для платформы Xbox One по 'user_score'.

Проверим на наличие пропущенных значений.

```python
xbox_one = u.loc[df['platform']== 'XOne', 'user_score']
print('Кол-во пропущенных значений', xbox_one.isna().sum())
display(xbox_one.describe())
```

Определим необходимые значения для платформы PC по 'user_score'.

Проверим на наличие пропущенных значений.

```python
pc = u.loc[df['platform']== 'PC', 'user_score']
print('Кол-во пропущенных значений', pc.isna().sum()) #посчитаем кол-во пропущенных значений
display(pc.describe())# выведем информацию о выборке
```

Т.к. выборки у вас разного размера и СКВО разные, то укажем параметр equal_var = False, и тест посчитает дисперсии выборок по отдельности.

```python
results = st.ttest_ind(xbox_one, pc, equal_var = False)# вызов метода для проверки гипотезы

alpha = 0.05# критический уровень статистической значимости

print('p-значение:', results.pvalue)# вывод значения p-value на экран

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: оценки пользователей на этих платформах отличаются')
else:
    print('Не получилось отвергнуть нулевую гипотезу: оценки пользователей на этих платформах не отличаются')
```

Значение p-value показывает: вероятность случайно получить такое или большее различие равна 41 %, т.к. принятый  критический уровень вероятности 5%, а получившееся значение более 5%, то гипотезу не следует отвергать как маловероятную. Средние оценки пользователей этих платформ не отличаются.

Построим гистограммы для наглядности

```python
fig, axes = plt.subplots()
axes = (u
 .query('platform == "XOne"')['user_score']
 .plot.hist(alpha=0.5, bins=20, range=(0,10))
)
axes = (u
 .query('platform == "PC"')['user_score']
 .plot.hist(alpha=0.5, bins=20, range=(0,10))
        )
axes.legend(["XOne", "PC"])
fig.suptitle('Распределение по рейтингу')
plt.show()
```
