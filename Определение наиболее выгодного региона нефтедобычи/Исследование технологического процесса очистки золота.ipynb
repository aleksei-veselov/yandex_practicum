# Описание данных
**Технологический процесс**

- *Rougher feed* — исходное сырье;
- *Rougher additions* (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant:

1) *Xanthate* — ксантогенат (промотер, или активатор флотации);

2) *Sulphate* — сульфат (на данном производстве сульфид натрия);

3) *Depressant* — депрессант (силикат натрия).

- *Rougher process* (англ. «грубый процесс») — флотация;
- *Rougher tails* — отвальные хвосты;
- *Float banks* — флотационная установка;
- *Cleaner process* — очистка;
- *Rougher Au* — черновой концентрат золота;
- *Final Au* — финальный концентрат золота.

**Параметры этапов**
- *air amount* — объём воздуха;
- *fluid levels* — уровень жидкости;
- *feed size* — размер гранул сырья;
- *feed rate* — скорость подачи.

**Наименование признаков**

*Наименование признаков должно быть такое:*

**[этап].[тип_параметра].[название_параметра]**

**Пример:** rougher.input.feed_ag

*Возможные значения для блока* **[этап]:**
- rougher — флотация;
- primary_cleaner — первичная очистка;
- secondary_cleaner — вторичная очистка;
- final — финальные характеристики.

*Возможные значения для блока* **[тип_параметра]:**
- *input* — параметры сырья;
- output* — параметры продукта;
- *state* — параметры, характеризующие текущее состояние этапа;
- *calculation* — расчётные характеристики.

**Данные находятся в трёх файлах:**

### Проверка правильности расчета эффективности обогащения

**Эффективность обогащения рассчитывается по формуле:**

**recovery** = ((C*(F-T))/(F*(C-T))*100%

где:
- C — доля золота в концентрате после флотации/очистки (rougher.output.concentrate_au);
- F — доля золота в сырье/концентрате до флотации/очистки (rougher.input.feed_au);
- T — доля золота в отвальных хвостах после флотации/очистки (rougher.output.tail_au).

Произведем расчет recovery на обучающей выборке для признака **rougher.output.recovery**

```python
C = df_train['rougher.output.concentrate_au']

F = df_train['rougher.input.feed_au']

T = df_train['rougher.output.tail_au']

#Проверка на наличие пропущенных значений
print(C.isna().sum())
print(F.isna().sum())
print(T.isna().sum())
```

```python
recovery = ((C*(F-T))/(F*(C-T)))*100
```

Найдем MAE между найденными значениями recovery на обучающей выборке для признака rougher.output.recovery и значением признака. 

```python
print('Значение MAE = ', mean_absolute_error(df_train['rougher.output.recovery'], recovery))
```

Значение MAE =  9.83758577644259e-15, говорит нам о том, что средняя разница между фактическим значением данных и значением, рассчитанными значениями мала, следовательно может быть применена в качестве целевого признака.

### Исследование изменения концентрации металлов (Au, Ag, Pb) на различных этапах ТП

```python
metal = {'ag':['rougher.input.feed_ag',
                     'rougher.output.concentrate_ag',
                     'primary_cleaner.output.concentrate_ag',
                     'final.output.concentrate_ag' ],
               'au':['rougher.input.feed_au',
                     'rougher.output.concentrate_au',
                     'primary_cleaner.output.concentrate_au',
                     'final.output.concentrate_au' ],
               'pb':['rougher.input.feed_pb',
                     'rougher.output.concentrate_pb',
                     'primary_cleaner.output.concentrate_pb',
                     'final.output.concentrate_pb' ]}
for i in metal:
    plt.figure(figsize=(19,10))
    sns.histplot(data = df_train[metal[i]])
    plt.title(f'График распределения изменений концентрации металлa {i} на различных этапах ТП', fontsize=20)
    plt.xlabel('Концентрация металла', fontsize=20)
    plt.ylabel('Количество',fontsize=20)
    plt.show()
```

- Концентрация золота(AU) увеличивается с первой стадии и до последней стадии ТП;

- Концентрация свинца(PB)  увеличивается на этапе флотации и после первичной очитски, затем остатется примерно на одном уровне

## Модель

### Функция для вычисления итоговой sMAPE

```python
def sMAPE(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

### Обучеине моделей

Функция для оценки качества модели

```python
def cross_val (model, features, target):
    scores = cross_val_score(
        model,
        features,
        target,
        scoring=make_scorer(sMAPE),
        cv=5)
    smape = pd.Series(scores).mean()
    return smape
```

Функция для вывода оценок качества модели

```python
def sample(model):
    sMAPE_rougher = cross_val (
        model,
        features_rougher_df_train,
        target_rougher_df_train
    )
    
    sMAPE_final = cross_val (
        model,
        features_final_df_train,
        target_final_df_train
    )
    
    sMAPE_end = 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
    print('Средняя оценка sMAPE на этапе rougher:', sMAPE_rougher)
    print('Средняя оценка sMAPE на финальном этапе:', sMAPE_final)
    print('Итоговое sMAPE:', sMAPE_end)
```

Модель **RandomForestRegressor**  с применением RandomizedSearchCV

```python
rfr_space={'max_depth': range(4, 40, 2), # ДО РЕВЬЮЕРА БЫЛО: range(10, 100, 30),
              'n_estimators':range(10, 150, 50),
              'min_samples_leaf':range(1,5),
              'min_samples_split':range(2, 10, 2)
          }

rfr = RandomForestRegressor(random_state=12345) # RANDOM STATE ОТ РЕВЬЮЕРА
rfr_random = RandomizedSearchCV(rfr, rfr_space, n_iter=5, scoring=make_scorer(sMAPE), cv=5, error_score='raise', verbose=3) # VERBOSE ОТ РЕВЬЮЕРА
model_rouger = rfr_random.fit(features_rougher_df_train, target_rougher_df_train)
model_final =  rfr_random.fit(features_final_df_train, target_final_df_train)
sMAPE_rougher = model_rouger.best_score_
sMAPE_final = model_final.best_score_
sMAPE_end = 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
print('Лучшие гиперпараметры model_rouger: '+str(model_rouger.best_params_))

print('Лучшие гиперпараметры model_final: '+str(model_final.best_params_))

print('Итоговое sMAPE:', sMAPE_end)
```
